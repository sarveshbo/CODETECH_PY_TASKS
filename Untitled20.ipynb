```python
import requests
import matplotlib.pyplot as plt
import datetime

# OpenWeatherMap API details
API_KEY = "547dae707c9da48b0e299bc6b6fbc3dc"  # Replace with your actual API key
LAT, LON = 33.44, -94.04  # Example location (Modify as needed)
URL = f"https://api.openweathermap.org/data/2.5/forecast?lat=33.44&lon=-94.04&appid=547dae707c9da48b0e299bc6b6fbc3dc&units=metric"

# Fetch data from API
response = requests.get(URL)
data = response.json()

# Extract relevant information
dates = []
temperatures = []
humidity = []

for entry in data["list"]:
    dt = datetime.datetime.fromtimestamp(entry["dt"])
    dates.append(dt)
    temperatures.append(entry["main"]["temp"])
    humidity.append(entry["main"]["humidity"])

# Plot temperature and humidity trends
plt.figure(figsize=(12, 5))
plt.plot(dates, temperatures, marker='o', label="Temperature (Â°C)", color='r')
plt.plot(dates, humidity, marker='s', label="Humidity (%)", color='b')
plt.xlabel("Date & Time")
plt.ylabel("Value")
plt.title("Weather Forecast: Temperature & Humidity Trends")
plt.legend()
plt.xticks(rotation=45)
plt.grid()

plt.show()

```
